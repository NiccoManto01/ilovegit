```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jackknife(values):
    mean = np.mean(values)        
    return mean - (values - mean)/(len(values) - 1)

def jackknife_var(cluster):
    return ((len(cluster)-1)/len(cluster))*np.sum(np.power(cluster - np.mean(cluster), 2)) 

def func_anomaly(values, t0, dim):
    return values*(t0**2)/(dim**4)
```

```python
   
#Apertura file Q_data.txt
data_12 = np.loadtxt('Grafici/run1_12_sums_Dani.txt')

# Plotting
fig, ax = plt.subplots(2, 2, figsize=(15,12))
fig.suptitle('Q data for run12 distribution for different values of Wilson time (tw)', fontsize=16)

ax[0][0].hist(data_12[0,:], bins=100, density=True, alpha=0.5, label='Q data_12')
ax[0][0].set_title('tw = 0')
ax[0][0].set_xlabel('Q')
ax[0][0].set_ylabel('Density')

ax[0][1].hist(data_12[3,:], bins=100, density=True, alpha=0.5, label='Q data_12')
ax[0][1].set_title('tw = 3')
ax[0][1].set_xlabel('Q')
ax[0][1].set_ylabel('Density')

ax[1][0].hist(data_12[7,:], bins=100, density=True, alpha=0.5, label='Q data')
ax[1][0].set_title('tw = 7')
ax[1][0].set_xlabel('Q')
ax[1][0].set_ylabel('Density')

ax[1][1].hist(data_12[28,:], bins=100, density=True, alpha=0.5, label='Q data')
ax[1][1].set_title('tw = 28')
ax[1][1].set_xlabel('Q')
ax[1][1].set_ylabel('Density')

cluster_12 = jackknife(np.power(data_12[28,:],2))
cluster_12_var = jackknife_var(cluster_12)
anomaly_12_cluster = func_anomaly(cluster_12, 2.79, 12)
anomaly_12 = np.mean(anomaly_12_cluster)
anomaly_12_var = jackknife_var(anomaly_12_cluster)
print(f"Anomaly for run12: {anomaly_12:.16f} Â± {np.sqrt(anomaly_12_var):.16f}")

```

```python
from sklearn.linear_model import LinearRegression

wilson_index= np.array((1/2.79, 1/3.78, 1/4.87, 1/6.2))
anomalies = np.array((anomaly_12, anomaly_16b, anomaly_16, anomaly_20))
errors = np.sqrt((anomaly_12_var, anomaly_16b_var, anomaly_16_var, anomaly_20_var))

plt.errorbar(wilson_index, anomalies, yerr = errors, fmt='None',  markersize=1, capsize= 3)
plt.scatter(wilson_index, anomalies, label='Anomaly points')

lin_reg, lin_cov = np.polyfit(wilson_index, anomalies, 1, w=1/errors, cov=True)
a, b = lin_reg
err_a, err_b = np.sqrt(np.diag(lin_cov))

x = np.linspace(0, 1/2.5, 100)
plt.plot(x, b + a*np.array(x), label='Linear fit', color='red')
plt.scatter(0, b, label='Xcont estimated', color='red')
plt.errorbar(0, b, yerr=err_b, fmt='None', markersize=1, capsize=3, color='red')
plt.scatter(0, 6.64e-04, label='Xcont target', color='green')
plt.errorbar(0, 6.64e-04, yerr=7e-06, fmt='None', markersize=1, capsize=3, color='green')
plt.title('Anomaly vs Wilson index')
plt.xlabel('a^2/t0^2')
plt.ylabel('a^2*X')
plt.legend()
#Chi Quadratic test
chi_squared = np.sum(((anomalies - (b + a*np.array(wilson_index))) / errors) ** 2)
chi_squared_dof = chi_squared/ (len(wilson_index) - 2)
print(f"Chi-squared: {chi_squared:.4f}, DoF: {len(wilson_index) - 2}, Reduced Chi-squared: {chi_squared_dof:.4f}")

print(f"{anomaly_12:.16f}, {anomaly_16b:.16f}, {anomaly_16:.16f}, {anomaly_20:.16f}")



#AGGIUNGI DEVIAZIONE DAL VALORE ATTESO
```

```python
#RIPETIAMO IL PROCESSO PER DIVERSI VALORI DI WILSON TIME

fig, ax = plt.subplots(figsize=(15, 10))

frazioni = np.array((0.25, 0.5, 0.75, 1.4))
frac_matrix = np.zeros((len(frazioni), 4))
errors_matrix = np.zeros((len(frazioni), 4))
lin_coeff = np.zeros((len(frazioni), 4))
cluster_original_12 = anomaly_12_cluster 
cluster_original_16b = anomaly_16b_cluster
cluster_original_16 = anomaly_16_cluster
cluster_original_20 = anomaly_20_cluster
colors = ['blue', 'orange', 'green', 'red']

for i,c, frac in zip(range(4),colors,frazioni):
    print(f"Processing fraction: {frac} t0, Wilson time:{int(frac*28)} for run 12")
    cluster_12 = jackknife(np.power(data_12[int(frac*28),:],2))
    cluster_12_var = jackknife_var(cluster_12)
    anomaly_12_cluster = func_anomaly(cluster_12, 2.8, 12)
    frac_12_cluster = anomaly_12_cluster/cluster_original_12
    frac_matrix[i,0] = np.mean(frac_12_cluster)
    errors_matrix[i,0] = np.sqrt(jackknife_var(frac_12_cluster))

    print(f"Processing fraction: {frac} t0, Wilson time:{int(frac*38)} for run 16b")
    cluster_16b = jackknife(np.power(data_16b[int(frac*38),:],2))
    cluster_16b_var = jackknife_var(cluster_16b)
    anomaly_16b_cluster = func_anomaly(cluster_16b, 3.8, 16)
    frac_16b_cluster = anomaly_16b_cluster/cluster_original_16b
    frac_matrix[i,1] = np.mean(frac_16b_cluster)
    errors_matrix[i,1] = np.sqrt(jackknife_var(frac_16b_cluster))

    print(f"Processing fraction: {frac} t0, Wilson time:{int(frac*49)} for run 16")
    cluster_16 = jackknife(np.power(data_16[int(frac*49),:],2))
    cluster_16_var = jackknife_var(cluster_16)
    anomaly_16_cluster = func_anomaly(cluster_16, 4.9, 16)
    frac_16_cluster = anomaly_16_cluster/cluster_original_16
    frac_matrix[i,2] = np.mean(frac_16_cluster)
    errors_matrix[i,2] = np.sqrt(jackknife_var(frac_16_cluster))

    print(f"Processing fraction: {frac} t0, Wilson time:{int(frac*62)} for run 20")
    cluster_20 = jackknife(np.power(data_20[int(frac*62),:],2))
    cluster_20_var = jackknife_var(cluster_20)
    anomaly_20_cluster = func_anomaly(cluster_20, 6.2, 20)
    frac_20_cluster = anomaly_20_cluster/cluster_original_20
    frac_matrix[i,3] = np.mean(frac_20_cluster)
    errors_matrix[i,3] = np.sqrt(jackknife_var(frac_20_cluster))

    lin_reg, lin_cov = np.polyfit(wilson_index, frac_matrix[i,:], 1, w=1/errors_matrix[i,:], cov=True)
    lin_coeff[i,0] = lin_reg[1]  # Intercept
    lin_coeff[i,1] = np.sqrt(lin_cov[1,1])  # Error on intercept
    lin_coeff[i,2] = lin_reg[0]  # Slope
    lin_coeff[i,3] = np.sqrt(lin_cov[0,0])  # Error on slope
    
    # Plotting the fractions
    ax.scatter(wilson_index, frac_matrix[i,:], label =f'Fraction {frac}t0', s=15, zorder =2, color = c)
    ax.errorbar(wilson_index, frac_matrix[i,:], yerr=errors_matrix[i,:], fmt='None', markersize=1, capsize=3, color='black', zorder=1)

    #Plotting the linear fit of the fractions
    ax.plot(x, lin_coeff[i,0] + lin_coeff[i,2] * np.array(x), label='Linear fit', color=c)
    ax.scatter(0, lin_coeff[i,0], color=c, s= 15, zorder=2)
    ax.errorbar(0, lin_coeff[i,0], yerr=abs(lin_coeff[i,1]), fmt='None', markersize=1, capsize=3, color=c)

ax.scatter(0, 1, label = 'Expected limit', color='black', s= 100, zorder=3, marker='*')
ax.legend()
ax.set_title('Anomaly fractions vs Wilson index for different fractions of t0')
ax.set_xlabel('a^2/t0^2')
ax.set_ylabel('Xt/X')
print("Linear coefficients for each fraction:")
print(lin_coeff, "\n")

print("Fraction matrix:")
print(frac_matrix, "\n")

print("Errors matrix:")
print(errors_matrix)


    
```
